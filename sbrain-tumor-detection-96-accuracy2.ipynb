```python
import streamlit as st
import tensorflow as tf
from tensorflow.keras.preprocessing import image
import numpy as np
from PIL import Image
```

```python
@st.cache_resource
def load_model():
    model = tf.keras.models.load_model('C:\\Users\\green\\AppData\\Local\\Microsoft\\Windows\\INetCache\\IE\\3J8XUUE7\\brain-tumor-detection-96-accuracy (2)[1].ipynb')
    return model
```

```python
categories = ["glioma", "meningioma", "no tumor", "pituitary"]
```

```python
st.title("Brain Tumor Detection")
st.write("Upload an MRI image to detect if there is a tumor and what type it is.")
```

```python
uploaded_file = st.file_uploader("Choose an MRI image...", type="jpg")
```

```python
if uploaded_file is not None:
    # Display the uploaded image
    img = Image.open(uploaded_file)
    st.image(img, caption='Uploaded MRI Image', use_column_width=True)
```

```python
img = img.resize((150, 150))  # Resize to match model input size
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array = img_array / 255.0
```

```python
prediction = model.predict(img_array)
predicted_class = np.argmax(prediction)

    
st.write(f"Prediction: {categories[predicted_class]}")
```
